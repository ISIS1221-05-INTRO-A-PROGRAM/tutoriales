# Parte 1: Usemos funciones de Python

## int() - Convertir a entero
La función **int()** permite convertir una variable en su representación entera. Por ejemplo: Convertir un texto a número, un número real a entero, etc.

## float() - Convertir a real
La función **float()** permite convertir una variable en un número real. Por ejemplo: Convertir un texto a real, un entero a real, etc.

## str() - Convertir a texto
La función **str()** permite convertir una variable en su representación en texto. Por ejemplo: Convertir un número entero o real a texto, convertir un valor de verdad a texto, etc.

## type() - Dar el tipo de una variable
La función **type()** permite consultar el tipo de una variable. Este valor se puede imprimir.

Veamos como funciona:

```python
# Esta instrucción convierte un texto que representa un entero en número entero y lo guarda en una variable
entero = int("-11")
print(type(entero))
# Esta segunda instrucción pide al usuario un número en texto y lo convierte a número real
real = input("Ingrese un número real")
print("Tipo antes de la conversión:",type(real))
# Esta instrucción convierte el texto solicitado en un número real
real = float(real)
print("Tipo después de la conversión:",type(real))

# Esta instrucción suma los dos números, el entero y el real
suma = entero+real
# Esta instrucción crea un texto concatenando el texto "El resultado es: " y la suma de los dos números en texto
resultado_texto = "El resultado es: "+str(suma)
# Esta instrucción muestra el resultado de la concatenación
print(resultado_texto)
```

# Parte 2: Usemos funciones aritméticas

## abs() - Valor absoluto
La función **abs()** permite calcular el valor absoluto de un número.

## round() - Redondeo
La función **round()** permite redondear un número a entero o a un real si se le especifican cuántas cifras decimales se necesitan.

## min() - Mínimo
La función **min()** permite convertir una variable en su representación en texto. Por ejemplo: Convertir un número entero o real a texto, convertir un valor de verdad a texto, etc.

## max() - Máximo
La función **max()** permite consultar el tipo de una variable. Este valor se puede imprimir.

Veamos como funciona:

```python
# Esta instrucción muestra el valor absoluto de un número
print("Prueba de valor absoluto",abs(-1.2))

# Esta instrucción redondea un número a entero
print("Prueba de redondeo a entero", round(3.141543))
print("Tipo de un redondeo entero", type(round(1.2)))

# Esta instrucción redondea un número real a dos cifras decimales
print("Prueba de redondeo a dos cifras decimales", round(3.141543,2))

# Esta instrucción da el máximo entre dos números
print("Prueba de máximo",max(1,10))

# Esta instrucción da el mínimo entre dos números
print("Prueba de mínimo",min(3,5))

```

# Parte 4 : Definición de funciones

## Sintaxis
Python permite que el usuario defina sus propias funciones para poder reutilizarlas. Esto se hace especificando:
* El nombre con el que será invocada la función
* Las entradas que tendrá, cuales y de qué tipo
* La salida que tendrá y de qué tipo


```python
# Definamos una función que calcule el perímetro de un rectángulo.
# Esta línea define una función , le asigna el nombre perimetro_rectangulo, 
# define dos entradas llamadas lado1 y lado2, ambos números reales y define una respuesta de tipo número real
def perimetro_rectangulo(lado1:float, lado2:float)->float:
    # Aquí va un bloque de instrucciones que resuelven el problema. 
    # Ese bloque de instrucciones es el cuerpo de la función
    # Se delimita usando una tabulación 
    perimetro = lado1*2 + lado2*2
    return perimetro

# usemos la función para calcular el perimetro de un rectángulo

l1 = float(input("Ingrese la longitud del primer lado"))
l2 = float(input("Ingrese la longitud del segundo lado"))
# En esta línea se está invocando/llamando a la función recién creada
print("El perímetro es:",perimetro_rectangulo(l1, l2))
```
