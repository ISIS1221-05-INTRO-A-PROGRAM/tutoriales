# Tutorial 5: Ciclos usando la instrucción while

Los ciclos en Python son una estructura de control que se utiliza para repetir instrucciones múltiples veces. Los ciclos son una herramienta especialmente útil para resolver problemas complejos. En este tutorial, aprenderás más sobre los ciclos en Python.





# Parte 1: Uso básico de ciclos con while

## Un primer ciclo de ejemplo: Contar de 1 a n 
Uno de los ciclos más sencillos que se puede escribir es uno que permita contar de 1 a n. 

Aquí tienes un ejemplo de cómo se usa:


```python
# En este ejemplo, la variable i se inicializa en 0 y se incrementa en 1 en cada iteración.
# El usuario ingresa un valor mayor a 0, especificando el numero de iteraciones que se harán.
i=0
n=int(input("Introduce un valor mayor a 0:"))
while i<n:
    print("Valor del conteo:",i+1)
    i+=1
```

## Contar de 1 a n pero de 2 en 2 
Vamos a modificar el ciclo anterior para hacer el proceso contando de dos en dos 

Aquí tienes un ejemplo de cómo se hace:

```python
# En este ejemplo, la variable i se inicializa en 0 y se incrementa en 2 en cada iteración.
# El usuario ingresa un valor mayor a 0, especificando el numero de iteraciones que se harán.
i=0
n=int(input("Introduce un valor mayor a 0:"))
while i<n:
    print("Valor del conteo:",i+2)
    i+=2
```

## Contar la cantidad de ocurrencias de una letra en una palabra
Para contar la cantidad de ocurrencias de una letra en una palabra, basta con recorrer la cadena y preguntar si la letra actual es la que esyou buscando.



```python
palabra=input("Ingrese una palabra: ")
letra=input("Ingrese una letra: ")

indice = 0
ocurrencias = 0
tamaño=len(palabra)
while indice <tamaño:
    actual = palabra[indice]
    
    if actual == letra:
        print("Revisando letra:", actual, " ocurrencia positiva")
        ocurrencias+=1
    else:
        print("Revisando letra:", actual, " ocurrencia negativa")
    indice+=1
print("La cantidad de veces que '", letra, "' está en '", palabra, "' es: ", ocurrencias, sep="")
```

## Contar la cantidad de ocurrencias de una letra en una palabra (otra versión)
En esta versión de cómo contar la cantidad de ocurrencias de una palabra en una letra iremos disminuyendo el tamaño de la palabra mediante un slicing. Así mismo, la condición del ciclo está escrita en términos del tamaño de lo que falta revisar.

```python
palabra=input("Ingrese una palabra: ")
letra=input("Ingrese una letra: ")
contador = 0

while len(palabra)>0:
    actual = palabra[0]
    print("Revisando letra: ", actual)
    if actual == letra:
        contador+=1
    palabra= palabra[1:]
    print("Falta por revisar: ", palabra)
print("La cantidad de veces que '", letra, "' está en '", palabra, "' es: ", contador, sep="")
```

## Sumar una secuencia de dígitos
Este programa usa suma la secuencia de dígitos que el usuario ingresa. 

Veamos cómo funciona:


```python
digitos=input("Escriba una secuencia de dígitos: ")
i =0
suma=0 
while i<len(digitos):
    digito= int(digitos[i])
    print("Dígito a evaluar: ", digito)
    suma+=digito
    print("Sumatoria hasta ahora: ",suma)
    i+=1
print("Total de la sumatoria: ",suma)
```

## Sencillo algoritmo para encriptar un texto

En este ejemplo vamos a encriptar un texto sumándole a su valor númerico la posición en la que está cada uno de los caracteres del texto.

Veámos cómo:


```python
frase=input("Escriba una frase: ")
i =0
salida="" 
while i<len(frase):
    letra= frase[i]
    print("Letra a encriptar: ", letra)

    val=ord(letra)
    val+=i
    letra_encriptada=chr(val)
    print("Letra encriptada: ", letra_encriptada)
    salida+= letra_encriptada

    i+=1
print("Frase encriptada: ",salida)
```

## Dibujar un arbol de navidad
En este sencillo ejemplo, dibujaremos un árbol de navidad en la consola

```python
i=1
while i < 30:
    print(('^'*i).center(30))
    i+=2

leg=0
while leg<4:
    print(('|||').center(30))
    leg+=1

print(('\_____/').center(30))
print(' '+30*'-')
print('(¨¨¨)¡Feliz Navidad! (¨¨¨)'.center(30))
```
