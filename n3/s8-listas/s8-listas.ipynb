## 1. Creación de listas
Las listas en Python se pueden crear utilizando corchetes []. Pueden contener cualquier tipo de datos, incluyendo otros tipos de listas.


```python
# Creación de una lista vacía
lista_vacia = []

# Creación de una lista con elementos
mi_lista = [1, 2, 3, 4, 5]
lista_de_listas = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
lista_de_diccionarios = [{"a": 1, "b": 2}, {"c": 3, "d": 4}]

print("Mi lista:", mi_lista)
print("Lista de listas:", lista_de_listas)
print("Lista de diccionarios:", lista_de_diccionarios)
```

## 2. Acceso a elementos
Puedes acceder a elementos individuales de una lista utilizando índices. Los índices comienzan desde 0.

```python
mi_lista = [1, 2, 3, 4, 5]

# Acceso a un elemento
primer_elemento = mi_lista[0]
segundo_elemento = mi_lista[1]

# Acceso a un rango de elementos
sublista = mi_lista[2:4]  # Accede a los elementos desde el índice 2 hasta el índice 3 (no incluido)

print("Primer elemento: ", primer_elemento)
print("Segundo elemento: ", segundo_elemento)
print("Sublista: ", sublista)
```

## 3. Modificación de elementos
Puedes modificar elementos individuales de una lista utilizando índices.



```python
mi_lista = [1, 2, 3, 4, 5]

# Modificar un elemento
mi_lista[0] = 10

# Modificar un rango de elementos
mi_lista[1:3] = [20, 30]

print("Mi lista: " , mi_lista)
```

## 4. Añadir y eliminar elementos
Puedes añadir y eliminar elementos de una lista utilizando métodos como **append()**, **insert()**, **extend()** y **remove()**.


```python
mi_lista = [1, 2, 3]

# Añadir un elemento al final
mi_lista.append(4)

# Insertar un elemento en una posición específica
mi_lista.insert(1, 5)

# Extender la lista con otra lista
mi_lista.extend([6, 7, 8])

# Eliminar un elemento
mi_lista.remove(3)

# Eliminar el último elemento
ultimo_elemento = mi_lista.pop()

print("Mi lista: ",mi_lista)
print("Último elemento: ",ultimo_elemento)
```

## 5. Operaciones comunes
Las listas en Python admiten varias operaciones comunes, como la concatenación, repetición y búsqueda de elementos.

```python
lista1 = [1, 2, 3]
lista2 = [4, 5, 6]

# Concatenación de listas
concatenada = lista1 + lista2

# Repetición de una lista
repetida = lista1 * 2

# Búsqueda de elementos
esta_en_la_lista = 2 in lista1

print("Lista concatenada: ", concatenada)
print("Lista repetida: ", repetida)
print("2 está en la lista: ", esta_en_la_lista)
```

## 6. Recorrido de listas con while
Las listas se pueden recorrer con un ciclo .

```python
mi_lista = [1,2,3,4,5,6,7,8,9]
sumatoria=0
i =0
while i<len(mi_lista):
    sumatoria += mi_lista[i]
    i+=1
print("Resultado de la sumatoria: ", sumatoria)
```

## 7. Recorrido de listas con for y range (por índices)
El ciclo for permite recorrer la lista . En este ejemplo lo haremos por índice usando la función **range()**

```python
mi_lista = [1,2,3,4,5,6,7,8,9]
sumatoria=0
for i in range(len(mi_lista)):
    sumatoria += mi_lista[i]
print("Resultado de la sumatoria: ", sumatoria)
```

## 8. Recorrido de listas con for (por elementos)
El ciclo for permite recorrer la lista . En este ejemplo lo haremos por elementos

```python
mi_lista = [1,2,3,4,5,6,7,8,9]
sumatoria=0
for elemento in mi_lista:
    sumatoria += elemento
print("Resultado de la sumatoria: ", sumatoria)
```
