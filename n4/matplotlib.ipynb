# Tipos de gráficos en Matplotlib

La librería Matplotlib permite hacer bastantes tipos de gráficos en 2d y 3d. Para mostrarlos, haremos un ejemplo con calificaciones de un examen y los visualizaremos de diferentes maneras. 
Primero generamos los datos:


```python
import matplotlib.pyplot as plt
import numpy as np


# Generar datos de ejemplo (notas aleatorias entre 0 y 100 para 30 estudiantes)
num_estudiantes = 30
np.random.seed(0)
notas = np.random.randint(0, 101, num_estudiantes)

# Crear una lista de identificadores de estudiante
estudiantes = ['Estudiante {}'.format(i+1) for i in range(num_estudiantes)]

# Generar datos de ejemplo para el eje y (posiciones aleatorias en el eje y)
posiciones_y = np.random.rand(num_estudiantes)
```

```python
# Generar datos de ejemplo (notas aleatorias entre 0 y 100 para 30 estudiantes)
num_estudiantes = 30
notas = np.random.randint(0, 101, num_estudiantes)

# Imprimir la tabla de notas
print("Tabla de Notas de Estudiantes")
print("---------------------------------------------")
print("Estudiante\t|\tNota")
print("---------------------------------------------")
for i, nota in enumerate(notas):
    print("Estudiante {}\t|\t{}".format(i+1, nota))
```

## Gráfico de barras:
El gráfico de barras permite mostrar los valores de cada estudiante, donde x es cada estudiante, y su calificación.

```python
# Crear el gráfico de barras
plt.figure(figsize=(10, 6))
plt.bar(estudiantes, notas, color='skyblue')

# Agregar etiquetas y título
plt.xlabel('Estudiantes')
plt.ylabel('Notas')
plt.title('Notas de Exámenes de 30 Estudiantes')

# Rotar las etiquetas del eje x para una mejor visualización
plt.xticks(rotation=45, ha='right')

# Mostrar la gráfica
plt.tight_layout()
plt.show()
```

## Gráfico de dispersión:
El gráfico de dispersión permite mostrar los valores de cada estudiante como una coordenada, donde x es cada estudiante, y su calificación.

```python

# Crear el gráfico de dispersión
plt.figure(figsize=(10, 6))
plt.scatter(estudiantes, notas, c=posiciones_y, cmap='coolwarm', s=100, alpha=0.7)

# Agregar etiquetas y título
plt.xlabel('Estudiantes')
plt.ylabel('Notas')
plt.title('Notas de Exámenes de 30 Estudiantes (Scatter Plot)')

# Rotar las etiquetas del eje x para una mejor visualización
plt.xticks(rotation=45, ha='right')

# Añadir barra de color
cbar = plt.colorbar()
cbar.set_label('Posición en el eje Y')

# Mostrar la gráfica
plt.tight_layout()
plt.show()
```

## Gráfico de líneas:
El gráfico de líneas permite graficar la tendencia a subir o bajar. En este caso x es cada estudiante, y su calificación.

```python
# Crear el gráfico de línea
plt.figure(figsize=(10, 6))
plt.plot(estudiantes, notas, marker='o', linestyle='-')

# Agregar etiquetas y título
plt.xlabel('Estudiantes')
plt.ylabel('Notas')
plt.title('Notas de Exámenes de 30 Estudiantes (Gráfico de Línea)')

# Rotar las etiquetas del eje x para una mejor visualización
plt.xticks(rotation=45, ha='right')

# Mostrar la gráfica
plt.tight_layout()
plt.grid(True)
plt.show()
```

## Gráfico de histograma:
El gráfico de histograma permite mostrar las frecuencias de ciertos rangos de notas de los estudiantes. El eje x muestra los rangos y el eje y la cantidad de veces.

```python
# Crear el histograma
plt.figure(figsize=(10, 6))
plt.hist(notas, bins=10, color='skyblue', edgecolor='black', alpha=0.7)

# Agregar etiquetas y título
plt.xlabel('Notas')
plt.ylabel('Frecuencia')
plt.title('Histograma de Notas de Exámenes')

# Mostrar la gráfica
plt.grid(True)
plt.tight_layout()
plt.show()
```

## Gráfico de cajas:
El gráfico de cajas permite mostrar la distribución de las notas de los estudiantes. Específicamente muestra:
 - El valor mínimo (línea de la izquierda)
 - La mediana (línea del centro)
 - El máximo (línea de la derecha)
 - El cuartil Q1 (Borde izquierdo de la caja)
 - El cuartíl Q3 (Borde derecho de la caja)


```python
# Crear el boxplot
plt.figure(figsize=(8, 6))
plt.boxplot(notas, vert=False)

# Agregar etiquetas y título
plt.xlabel('Notas')
plt.title('Boxplot de Notas de Exámenes')

# Mostrar la gráfica
plt.grid(True)
plt.tight_layout()
plt.show()
```

## Gráfico de pie:
El gráfico de pie permite mostrar las proporción de ciertos valores categóricos. En este caso, se ve el porcentaje de estudiantes que aprobaron o perdieron.

```python
# Contar el número de estudiantes aprobados y reprobados
aprobados = np.sum(notas >= 60)
reprobados = np.sum(notas < 60)

# Crear un gráfico de pastel
plt.figure(figsize=(8, 6))
labels = ['Aprobados', 'Reprobados']
sizes = [aprobados, reprobados]
colors = ['lightgreen', 'salmon']
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140)

# Agregar título
plt.title('Proporción de Estudiantes Aprobados y Reprobados')

# Mostrar la gráfica
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.tight_layout()
plt.show()
```

## Gráfico de histograma 2d:
El gráfico de histograma 2D permite mostrar las frecuencias de ciertos rangos de notas de los estudiantes y de sus horas de estudio. Los ejes x y y muestran los rangos de nota y los rangos de horas de sueño, el color muestra la frecuencia (entre más oscuro, más frecuente).

```python
horas_estudio = np.random.randint(0, 30, num_estudiantes)  # Horas de estudio semanales (0 a 30 horas)

# Crear histograma 2D
plt.figure(figsize=(10, 8))
plt.hist2d(notas, horas_estudio, bins=(10, 10), cmap='Blues')

# Agregar etiquetas y título
plt.xlabel('Notas')
plt.ylabel('Horas de Estudio Semanales')
plt.title('Histograma 2D: Notas vs Horas de Estudio')

# Agregar barra de color
cbar = plt.colorbar()
cbar.set_label('Frecuencia')

# Mostrar la gráfica
plt.grid(True)
plt.tight_layout()
plt.show()
```

## Gráfico de dispersión 3D:
El gráfico de dispersión 3D permite ubicar los datos en un espacio 3D. En este caso, se ubican tres coordenadas: Notas, Horas de estudio y Horas de descanso

```python

from mpl_toolkits.mplot3d import Axes3D


horas_descanso = np.random.randint(5, 12, num_estudiantes)  # Horas de descanso diarias (5 a 12 horas)

# Crear figura 3D
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# Crear el scatter plot 3D
ax.scatter(notas, horas_estudio, horas_descanso, c='skyblue', marker='o')

# Agregar etiquetas y título
ax.set_xlabel('Notas')
ax.set_ylabel('Horas de Estudio Semanales')
ax.set_zlabel('Horas de Descanso Diarias')
ax.set_title('Scatter Plot 3D: Notas vs Horas de Estudio vs Horas de Descanso')

# Mostrar la gráfica
plt.tight_layout()
plt.show()
```
