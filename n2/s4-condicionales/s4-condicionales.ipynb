## Parte 1: if simple
La instrucción **if** permite crear una condición, la cual si se cumple ejecutará las instrucciones contenidas en ese condicional.

El flujo del programa se altera de la siguiente manera:

![Flujograma](https://drive.google.com/thumbnail?id=1Cp328uYETPMiAM4fTNyAa9Sf8AU_OlHe&sz=w300)

Veamos como funciona:

```python
# pidamos al usuario su edad y validaremos si es mayor de edad
edad = int(input("Ingrese su edad"))
# lo primero es usar la palabra if seguida de una expresión booleana y luego dos puntos
if edad >= 18:
    #Las instrucciones que están "dentro" se ejecutarán solo si la expresión del if es verdadera
    print("Usted es mayor de edad")

# Una vez superada la instrucción condicional, se retoma el flujo normal del programa
print("Validación finalizada")
```

## Parte 2: if en secuencia

Si se usan varias instrucciones **if** seguidas, se condicionarán varias instrucciones en secuencia. Cada bloque de instrucciones será independiente.

![Flujograma2](https://drive.google.com/thumbnail?id=1zquqFIqLEqBE7tdGP9TTc_oIbNczAoUv&sz=w300)

Veamos cómo funciona:

```python
# Supongamos un curso donde se realizan las siguientes aproximaciones en la nota definitiva:
# - aproximan a 5.0 las calificaciones iguales o superiores a 4.75
# - aproximan a 3.0 las calificaciones superiores a 2.75 y menores a 3.0
# - aproximan a 1.5 las calificaciones menores a 1.5
# - cualquier otra calificación no será aproximada
# Además, el profesor del curso hizo un bono sobre la nota definitiva que aplicará de la siguiente manera:
# - Si la definitiva con aproximación fué menor a 3.0, no aplica
# - Si la definitiva con aproximación es mayor sumada con el bono supera 5, no aplica (porque el estudiante no puede sacar más de 5.0)

definitiva = float(input("Ingrese su calificación: "))
bono = float(input("Ingrese el valor del bono"))

if definitiva >= 4.75:
    definitiva = 5.0
    print("La definitiva fué ajustada a 5.0")

if definitiva >2.75 and definitiva <3:
    definitiva = 3.0
    print("La definitiva fué ajustada a 3.0")

if definitiva < 1.75:
    definitiva = 1.5
    print("La definitiva fué ajustada a 1.5")

if definitiva >= 3 and (definitiva + bono) <= 5:
    definitiva += bono
    print("La bonificación fué aplicada")

print("El valor final fué: ", definitiva)
```

## Parte 3: elif
La instrucción **elif** permite especificar casos adicionales en un bloque condicional. Estos casos modelan situaciones mutuamente excluyentes que pueden suceder solamente una a la vez. Al darse uno de los casos especificados, los demás casos no serán considerados.


![Flujograma3](https://drive.google.com/thumbnail?id=1QE8b5iBqE3vZQqtQxiMMYGYS9mrx5naC&sz=w300)

Veamos como funciona:



```python
# Supongamos un juego de pirinola donde la pirinola puede tener los siguientes valores:

# - "TOMA 1: El jugador toma $100 de la mesa"
# - "TOMA 2: El jugador toma $200 de la mesa"
# - "TOMA TODO: El jugador toma todo el dinero de la mesa"
# - "PON 1: El jugador pone $100 en la mesa"
# - "PON 2: El jugador pone $200 en la mesa"
# - "TODOS PONEN: Todos los jugadores ponen $100 en la mesa"


# Programemos un algoritmo que calcule el valor de la mesa en una jugada

# Paso 1: generemos aleatoriamente un valor
import random
valor_pirinola = random.choice(["TOMA 1","TOMA 2", "TOMA TODO", "PON 1", "PON 2", "TODOS PONEN" ])
print("El valor de la pirinola fué: ",valor_pirinola)
# Paso 2: probemos con un valor inicial de 400
mesa = 400

# Paso 3: Usemos el if-elif para evaluar los casos
if valor_pirinola == "TOMA 1":
    mesa-=100
    print("El jugador ganó $100")
elif valor_pirinola == "TOMA 2":
    mesa-=200
    print("El jugador ganó $200")
elif valor_pirinola == "TOMA TODO":
    mesa=0
    print("El jugador ganó todo")
elif valor_pirinola == "PON 1":
    mesa+=100
    print("El jugador perdió $100")
elif valor_pirinola == "PON 2":
    mesa+=200
    print("El jugador perdió $200")
elif valor_pirinola == "TODOS PONEN":
    mesa+=400
    print("Todos los jugadores pierden $100")
print("La mesa quedó con: $",mesa, sep="")
```

## Parte 4: else
La instrucción **else** permite especificar un bloque de instrucciones adicional en un bloque condicional. Estas instrucciones se ejecutarán si ninguna de las condiciones del bloque if / elif se cumplen. Funciona como un "En cualquier otro caso".


![Flujograma4 else](https://drive.google.com/thumbnail?id=1di7zBCsJD0bjYCRFfgctMsLKlFJMPlgR&sz=w300)

Veamos como funciona:


```python
# Supongamos el diseño de un menú de opciones. Las opciones son las siguientes:
#  - Si el usuario selecciona la opción 1, Iniciará una partida
#  - Si el usuario selecciona la opción 2, Hará una jugada
#  - Si el usuario selecciona la opción 3, Terminará la partida
#  - Si el usuario selecciona la opción 4, Saldrá del juego
#  - Si el usuario selecciona cualquier otra opción, se le indicará que es incorrecta
print("Menu principal\n")
print("1. Nuevo juego")
print("2. Jugar")
print("3. Terminar juego")
print("4. Salir\n")
opcion = int(input("Seleccione la opción: "))

if opcion == 1:
    print("Se seleccionó la opción nuevo juego")
elif opcion == 2:
    print("Se seleccionó la opción jugar")
elif opcion == 3:
    print("Se seleccionó la opción terminar juego")
elif opcion == 4:
    print("Se seleccionó la opción salir")
else:
    print("Opción inválida")
```

## Parte 5: Bloques condicionales anidados

Un bloque de instrucciones condicionales puede tener otro bloque de condiciones anidadas. Esto representa subcasos de un caso principal.

![Flujograma5 anidamiento](https://drive.google.com/thumbnail?id=13tikQMcmAIBVO1e4uCRcCLO539FbXAsO&sz=w800)

Veamos cómo funciona:


```python
# Supongamos un juego de parqués donde se lanzan dos dados. De acuerdo al estado del juego, se puede.
# 1. Si el lanzamiento es doble :
#   1.1. Si es un doble 6 o un doble 1
#       1.1.1. Si tiene 2, 3 o 4 fichas en la carcel: Sacar todas las fichas
#       1.1.2. Si tiene 1 ficha en la carcel: Sacar la ficha y mover otra ficha
#       1.1.3. Si no tiene dichas en la carcel: Mover
#   1.2. Si es otro par
#       1.2.1. Si tiene 2,3 o 4 fichas en la carcel: Sacar 2 fichas
#       1.2.2. Si tiene 1 ficha en la carcel: Sacar la ficha y mover otra ficha
#       1.2.4 Si no tiene fichas en la carcel: Mover
# 2. Si el lanzamiento no es doble :
#   2.1. Si tiene todas las fichas en la carcel: 
#       2.1.1. Es la primera o la segunda vez que lanzamiento no es doble : Lanzar de nuevo
#       2.1.2. Es la tercera vez que lanzamiento no es doble: Terminar turno
#   2.2. No tiene todas las fichas en la carcel: Mover
# Nótese que los casos son un árbol de decisión

import random
# Paso 1: Simulemos el lanzamiento de dados
dado1= random.randint(1,6)
dado2= random.randint(1,6)
print("El valor de los dados es ","["+str(dado1)+","+str(dado2)+"]")
# Paso 2: Preguntemos al usuario cuántas fichas tiene en la carcel y qué número de lanzamiento es
carcel = int(input("¿Cuántas fichas tiene en la carcel?"))
veces = int(input("¿Cuántas veces ha lanzado los dados en este turno?"))

# Paso 3: Modelemos las situaciones

# Caso 1. Lanzamiento doble
if dado1==dado2:
    # Caso 1.1 Es doble 6 o doble 1
    if dado1== 6 or dado1== 1:
        # Caso 1.1.1 Si tiene 2, 3 o 4 fichas en la carcel: 
        if carcel > 1:
            # Sacar todas las fichas
            print("Debería sacar todas las fichas de la cárcel")
        # Caso 1.1.2. Si tiene 1 ficha en la carcel: 
        elif carcel == 1:
            # Sacar la ficha y mover otra ficha
            print("Debería sacar una ficha y mover otra ficha")
        # Caso 1.1.3. Si no tiene fichas en la carcel:
        else:
            # Mover
            print("Debería mover")
    # Caso 1.2. Si es otro par
    else:
        # Caso 1.2.1 Si tiene 2, 3 o 4 fichas en la carcel: 
        if carcel > 1:
            # Sacar 2 las fichas
            print("Debería sacar 2 fichas de la cárcel")
        # Caso 1.2.2. Si tiene 1 ficha en la carcel: 
        elif carcel == 1:
            # Sacar la ficha y mover otra ficha
            print("Debería sacar una ficha y mover otra ficha")
        # Caso 1.2.3. Si no tiene fichas en la carcel:
        else:
            # Mover
            print("Debería mover")
# Caso 2. Si el lanzamiento no es doble :
else:
    # Caso 2.1. Si tiene todas las fichas en la carcel:
    if carcel == 4:
        # Caso 2.1.1. Es la primera o la segunda vez que lanzamiento no es doble:
        if veces <3:
            # Lanzar de nuevo
            print("Puede lanzar de nuevo")
        # Caso 2.1.2. Es la tercera vez que lanzamiento no es doble:
        elif veces==3:
            # Terminar turno
            print("Su turno ha terminado")
    # Caso 2.2. No tiene todas las fichas en la carcel:
    else: 
        # Mover
        print("Debería mover")
        

```
